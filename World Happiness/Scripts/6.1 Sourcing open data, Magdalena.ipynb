# Create a financial status column

Highest GDP is 2.10, lowest is 0. I divided it into 3 categories (0-0.7, 0.7-1.4, and 1.4-2.1), and assigned it different walues. 

```python
def Financial_label (row):

  if row['Economy (GDP per Capita)'] <= 0.7:
    return 'Financially struggling'
  elif (row['Economy (GDP per Capita)'] > 0.7) and (row['Economy (GDP per Capita)'] < 1.4):
    return 'Financially average'
  elif row['Economy (GDP per Capita)'] >= 1.4:
    return 'Financially well-off'
  else: return 'Not enough data'
```

```python
#apply the function 
df_combined['Financial status'] = df_combined.apply(Financial_label, axis=1)
```

# Create Government status column

```python
def Government_label (row):

  if row['Trust (Government Corruption)'] <= 0.2:
    return 'Low government trust'
  elif (row['Trust (Government Corruption)'] > 0.2) and (row['Trust (Government Corruption)'] < 0.4):
    return 'Average government trust'
  elif row['Trust (Government Corruption)'] >= 0.4:
    return 'High government trust'
  else: return 'Not enough data'
```

```python
#apply the function 
df_combined['Government status'] = df_combined.apply(Government_label, axis=1)
```
